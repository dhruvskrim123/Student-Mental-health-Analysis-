```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Student Mental health.csv')
```

**Understanding the data**

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(f"Number of duplicate rows: {df.duplicated().sum()}")
```

**Clean the Data**

```python
# Remove duplicate rows
df = df.drop_duplicates()

```

```python
# Corrected list of categorical columns
categorical_cols = ['Choose your gender', 'Your current year of Study', 'Marital status',
                    'Do you have Depression?', 'Do you have Anxiety?',
                    'Do you have Panic attack?', 'Did you seek any specialist for a treatment?']
```

```python
# Convert them to 'category' type
for col in categorical_cols:
    df[col] = df[col].astype('category')

# Verify
print(df.info())

```

**Visualizing various mental health indicators**

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Do you have Depression?', data=df)
plt.title('Depression Distribution')
plt.xlabel('Depression')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Do you have Anxiety?', data=df)
plt.title('Anxiety Distribution')
plt.xlabel('Anxiety')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Do you have Panic attack?', data=df)
plt.title('Panic Attack Distribution')
plt.xlabel('Panic Attack')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Choose your gender', hue='Do you have Depression?', data=df)
plt.title('Depression by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Choose your gender', hue='Do you have Depression?', data=df)
plt.title('Depression by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Choose your gender', hue='Do you have Anxiety?', data=df)
plt.title('Anxiety by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Choose your gender', hue='Do you have Panic attack?', data=df)
plt.title('Panic Attack by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Convert Yes/No to 1/0
binary_cols = ['Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']
df[binary_cols] = df[binary_cols].replace({'Yes':1, 'No':0})

# Correlation matrix
corr_matrix = df[binary_cols].corr()
print(corr_matrix)

# Heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Between Mental Health Indicators')
plt.show()
```
